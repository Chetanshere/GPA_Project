#  Changing The Datatypes

```python
df1.info()
```

```python
df1['Reviews']=df1['Reviews'].astype(float)
```

```python
df1['Installs'] = df1['Installs'].str.replace('[^\d.]', '', regex=True)
```

```python
df1['Installs']= df1['Installs'].astype(float)
```

```python
df1['Price']=df1['Price'].str.replace('[^\d.]','',regex=True)
```

```python
df1['Price']=df1['Price'].astype(float)
```

```python
pd.to_datetime(df1['Last Updated'])
```

```python
df1.info()
```
